```python
import pandas as pd # 판다스 라이브러리 사용
import numpy as np # 넘파이 라이브러리 사용
import matplotlib.pyplot as plt # 그래프 출력 라이브러리
from sklearn.cluster import KMeans # k-평균 군집 라이브러리
# house.csv 파일을 다운로드
!wget -O house.csv "https://github.com/2022infotextbook/ai-basic/raw/main/2_ai_ml/house.csv"
# 다운로드한 house.csv 파일을 데이터 프레임으로 불러온다.
df = pd.read_csv("house.csv")
X = df.drop(columns=["House"]) # 데이터 전처리(정답 데이터 삭제)
model = KMeans (n_clusters= 4 , random_state=42) # k-평균 클러스터링
model.fit (X)
df["cluster"] = model.labels_ # 클러스터 할당
print(model.cluster_centers_ ) # 학습한 군집들의 중심 위치
plt.figure(figsize=(12, 8)) # 결과 시각화
plt.scatter(df["Ambition"], df["Courage"], c=df["cluster"], cmap="viridis")
plt.xlabel("Ambition")
plt.ylabel("Courage")
plt.title("k-Means: Ambition - Courage")
plt.show()
plt.figure(figsize=(12, 8)) # 결과 시각화
plt.scatter(df["Temperament"], df["Knowledge"], c=df["cluster"], cmap="viridis")
plt.xlabel("Temperament")
plt.ylabel("Knowledge")
plt.title("k-Means: Temperament - Knowledge")
plt.show()
new_data = np.array([[95, 70, 75, 85, 40]]) # 새로운 데이터에 대한 예측
prediction = model.predict (new_data)
print("용기95, 지식70, 호기심75, 성격85, 욕심40인 학생의 군집-> ", prediction[0])
```
