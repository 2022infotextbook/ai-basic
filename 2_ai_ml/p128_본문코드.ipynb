```python
!pip install SpeechRecognition # 음성 인식, 파일 변환에 필요한 라이브러리 설치
!apt-get install –y ffmpeg
import librosa.display # 시각화, 음성 인식, 파일 입출력에 필요한 라이브러리 가져오기
import IPython.display as ipd
import matplotlib.pyplot as plt
import speech_recognition as sr
import os
from google.colab import files
filename = list(files.upload().keys())[0] # 음성 파일 업로드, 업로드할 mp3 파일을 선택
plt.style.use('seaborn-white') # 시각화 스타일 지정
fig = plt.figure(figsize=(14,4)) # 업로드한 음성 파일 시각화
korean_wav, rate = librosa.core.load(filename)
librosa.display.waveshow(korean_wav, sr=rate)
ipd.Audio(korean_wav, rate=rate)
# mp3 파일을 wav 파일로 변환
os.system(f'ffmpeg -y -i "{filename}" -f wav -acodec pcm_s16le -ar 22050 -ac 1 output.wav')
r=sr.Recognizer() # 음성 인식 및 텍스트 변환
korean_audio = sr.AudioFile('output.wav')
with korean_audio as source:
 audio = r.record(source)
print(r.recognize_google(audio_data= audio, language='ko-KR')) # 한국어: ko-KR
```
